```python
import pandas as pd
import spacy
```

```python
df_gorokhov = pd.read_csv("gorokhov.csv", sep="\t")
df_gorokhov
```

```python
df_sovback = pd.read_csv("sovback.csv", sep="\t")
df_sovback
```

```python
df_syrnik = pd.read_csv("syrnik.csv", sep="\t")
df_syrnik
```

```python
nlp = spacy.load('ru_core_news_sm')
```

```python
def convert(df): #переводим тексты в формат списков со списками
  corp_1 = df["result"].tolist()
  corp = []
  for text in corp_1:
    text_1 = text[2:-2]
    text_2 = text_1.split("', '")
    corp.append(text_2)

  return corp

```

```python
gorokhov = convert(df_gorokhov)
syrnik = convert(df_syrnik)
sovback = convert(df_sovback)

print(sovback[:5])
```

```python
def add_metadata(corp): #добавляем метаинформацию - лемму и POS-tag
  corp_inf = []
  for text in corp:
    text_inf = []
    for sent in text:
      sent_inf = []
      doc = nlp(sent)
      for word in doc:
        if word.pos_ != 'PUNCT':
          word_inf = []
          word_inf.append(word)
          word_inf.append(word.lemma_)
          word_inf.append(word.pos_)
          sent_inf.append(word_inf)
      
      text_inf.append(sent_inf)

    corp_inf.append(text_inf)

  return corp_inf   
```

```python
meta_gorokhov = add_metadata(gorokhov)
for i in meta_gorokhov[:3]:
  print(i)
```

```python
meta_syrnik = add_metadata(syrnik)
for i in meta_syrnik[:3]:
  print(i)
```

```python
meta_sovback = add_metadata(sovback)
for i in meta_sovback[:3]:
  print(i)
```

```python
df_gorokhov['metadata'] = meta_gorokhov
df_syrnik['metadata'] = meta_syrnik
df_sovback['metadata'] = meta_sovback
```

```python
df_gorokhov.to_csv("metagorokhov.csv", index=None, sep="\t")
df_syrnik.to_csv("metasyrnik.csv", index=None, sep="\t")
df_sovback.to_csv("metasovback.csv", index=None, sep="\t")
```
